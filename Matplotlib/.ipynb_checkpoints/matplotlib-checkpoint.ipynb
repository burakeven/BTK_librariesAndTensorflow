```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline # bu kod jupitere ozeldir. Devamli plt.show yazmamiza gerek kalmaz.
```

```python
yasList = [10,20,30,40,50,60,70,20,30,10]
```

```python
kiloList = [20,60,80,75,43,45,23,56,77,35]
```

```python
numpyYasList = np.array(yasList)
```

```python
numpyKiloList = np.array(kiloList)
```

```python
plt.plot(numpyYasList, numpyKiloList, "g"); #burada g green anlamina gelir, yesil cizgiyle gosterir.
plt.xlabel("yas")
plt.ylabel("kilo")
```

```python
numpyArray1 = np.linspace(0,10,20)
```

```python
numpyArray1
```

```python
numpyArray2 = numpyArray1**2
```

```python
numpyArray2
```

```python
plt.plot(numpyArray1,numpyArray2,"g*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyArray1,numpyArray2,"r*-")
plt.subplot(1,2,2) #1 sira, 2 kolon ve su an ikinci grafigi ciziyoruz.
plt.plot(numpyArray2,numpyArray1, "b--")
```

```python
plt.plot(numpyArray1,numpyArray2,"r*-")
plt.plot(numpyArray2,numpyArray1, "b--")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.4,0.4,0.5,0.5]) #ic noktasi, 0.9 genislik ve uzunluk
figureAxes.plot(numpyArray1,numpyArray2, "g")
figureAxes.set_xlabel("X ekseni veri ismi")
figureAxes.set_ylabel("Y ekseni veri ismi")
figureAxes.set_title("Grafik Basligi")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.3,0.3,0.3,0.3])

eksen1.plot(numpyArray1, numpyArray2, "g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana grafik baslik")

eksen2.plot(numpyArray2, numpyArray1, "b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Kucuk grafik baslik")
```

```python
(benimFigure, benimEksen) = plt.subplots(nrows=1, ncols=2 )
type(benimEksen)

for eksen in benimEksen:
    eksen.plot(numpyArray1, numpyArray2, "g")
    eksen.set_xlabel("X ekseni")
plt.tight_layout() #iki grafik arasindaki mesafeyi ayarlar.
```

```python
yeniFigur = plt.figure() #burada figure parantez icerisine figsize dersek x ve y ekseni boyutlari ayarlanir.
#figure icerisine dpi=100 yazar isek bu da cozunurlugu ayarlar.
yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyArray1,numpyArray2, label="numpyArray ** 2")
yeniEksen.plot(numpyArray1, numpyArray1**3, label="numpyArray **3")
yeniEksen.legend(loc=2)
```

```python
yeniFigur.savefig("myFigure.png", dpi=200) #yukaridaki figuru kaydeder.
```

```python
numpyDizisi1 = np.linspace(0,10,20)
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1**2
```

```python
(benimFigur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2, "g", alpha=0.3) #saydamlik katmak istersek alpha degerini dusururuz.
benimEksen.plot(numpyDizisi2, numpyDizisi1, "b")
```

```python
(yeniFigur, yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 2, color="blue", linewidth = 4)
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 3, color="green")

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 4, color = "black", linestyle="--")

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 7, color = "red", linestyle="--", marker="o",markerfacecolor="black")
```
