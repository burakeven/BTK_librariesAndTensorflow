```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline #bu kod jupitere ozeldir. Devamli plt.show yazmamiza gerek kalmaz.
```

```python
yasList = [10,20,30,40,50,60,70,20,30,10]
```

```python
kiloList = [20,60,80,75,43,45,23,56,77,35]
```

```python
numpyYasList = np.array(yasList)
```

```python
numpyKiloList = np.array(kiloList)
```

```python
plt.plot(numpyYasList, numpyKiloList, "g"); #burada g green anlamina gelir, yesil cizgiyle gosterir.
plt.xlabel("yas")
plt.ylabel("kilo")
```

```python
numpyArray1 = np.linspace(0,10,20)
```

```python
numpyArray1
```

```python
numpyArray2 = numpyArray1**2
```

```python
numpyArray2
```

```python
plt.plot(numpyArray1,numpyArray2,"g*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyArray1,numpyArray2,"r*-")
plt.subplot(1,2,2) #1 sira, 2 kolon ve su an ikinci grafigi ciziyoruz.
plt.plot(numpyArray2,numpyArray1, "b--")
```

```python
plt.plot(numpyArray1,numpyArray2,"r*-")
plt.plot(numpyArray2,numpyArray1, "b--")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.4,0.4,0.5,0.5]) #ic noktasi, 0.9 genislik ve uzunluk
figureAxes.plot(numpyArray1,numpyArray2, "g")
figureAxes.set_xlabel("X ekseni veri ismi")
figureAxes.set_ylabel("Y ekseni veri ismi")
figureAxes.set_title("Grafik Basligi")
```
